```python
##Question 4:list of 10 countries with the highest number of kills?
nkill_in_countries= df.groupby('country_txt').max()['nkill'].nlargest(10)
nkill_in_countries
```

```python
plt.figure(figsize = (20,8))
x = range(len(nkill_in_countries))
ax = plt.axes();
ax.bar(x,nkill_in_countries, color=["#ff4dff"], width=0.8,label="nkill")
plt.title("list of 10 countries with the highest number of kills",fontsize=16)
plt.xticks(x,nkill_in_countries.index, rotation=25)
plt.xlabel("Different_Countries", fontsize=14, )
plt.ylabel("Number_of_Kills", fontsize=14)

for x in ax.patches:
    w=x.get_width()
    h=x.get_height()
    x,y=x.get_xy()
    
    ax.annotate(h, (x+ w/2, y+ h*1.02))
```

```python
##Question 5:list of 10 countries with the highest number of wounded?
nwound_in_countries= df.groupby('country_txt').max()['nwound'].nlargest(10)
nwound_in_countries
```

```python
plt.figure(figsize = (20,8))
x = range(len(nwound_in_countries))
ax = plt.axes();
ax.bar(x,nwound_in_countries, color=["#cc9900"], width=0.8,label="nwound")
plt.title("list of 10 countries with the highest number of kills",fontsize=16)
plt.xticks(x,nwound_in_countries.index, rotation=0)
plt.xlabel("Different_Countries", fontsize=14, )
plt.ylabel("Number_of_Wounded", fontsize=14)

for x in ax.patches:
    w=x.get_width()
    h=x.get_height()
    x,y=x.get_xy()
    
    ax.annotate(h, (x+ w/2, y+ h*1.02))
```
